```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "train.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "lespin/house-prices-dataset",
  file_path,
  # Provide any additional arguments like 
  # sql_query or pandas_kwargs. See the 
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", df.head())
```

```python
# Importando bibliotecas principais
import pandas as pd              # Trabalha com dados em formato de tabela (tipo planilhas)
import numpy as np               # Faz cálculos numéricos e trabalha com vetores e matrizes
import matplotlib.pyplot as plt  # Cria gráficos e visualizações
import seaborn as sns            # Cria gráficos mais bonitos e automáticos (baseado no Matplotlib)

# Importando ferramentas do scikit-learn
from sklearn.model_selection import train_test_split   # Divide os dados em treino e teste
from sklearn.linear_model import LinearRegression       # Modelo de Regressão Linear (usado para prever números)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score  # Mede o desempenho do modelo

# Ferramentas para preparar e transformar os dados
from sklearn.preprocessing import StandardScaler        # Padroniza os dados numéricos (média 0, desvio padrão 1)
from sklearn.preprocessing import OneHotEncoder         # Transforma textos em números binários (0 e 1)
from sklearn.preprocessing import LabelEncoder          # Transforma textos em números simples (ex: "Azul"=0, "Verde"=1)
```

```python
df = df.drop(columns=['Id','Alley','PoolQC','MasVnrType','MiscFeature','Fence','FireplaceQu'])
```

```python
# Preenche valores ausentes (NaN) com a mediana das colunas numéricas
df.fillna(
    df.median(numeric_only=True),  # Calcula a mediana apenas das colunas numéricas
    inplace=True                   # Altera o DataFrame original diretamente
)
df.fillna('None', inplace=True)  # Substitui os valores vazios por 'None' (uma string) e adiciona no DataFrame original

# Seleciona apenas as colunas do tipo "objeto" (normalmente textos ou categorias)
# É como se criasse uma lista com os nomes dessas colunas dentro do DataFrame
col_cats = df.select_dtypes(include=[object]).columns

# Para cada coluna de texto encontrada...
for col in col_cats:
    le = LabelEncoder()                 # Cria um codificador LabelEncoder
    df[col] = le.fit_transform(df[col]) # Converte os textos em números (ex: "Azul"→0, "Verde"→1, "Vermelho"→2). Sem isso, não teria como fazer regressão
```

```python
X = df.drop(columns=['SalePrice']) ## Tirou a variável SalePrice pois é justamente ela que queremos descobrir. Então X = todos os dados menos essas duas variáveis
y = df ['SalePrice']

## Separar dados de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.25, random_state=42)
```

# Rede Neural

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import numpy as np

```

```python
#set random random seeds for reproductibility
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

#contruindo o modelo
model = Sequential()
model.add(Dense(73, activation='relu', input_shape=(X_train.shape[1],))) #Camada de entrada com 73 neurônios (73 é o número de variáveis a serem avaliadas) e função de ativação Relu. é a camada que recebe os dados de entrada
model.add(Dense(37, activation='relu')) #Camada oculta com 37 neurônios e função de ativação Relu (é +- a metade)
model.add(Dense(19, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(1, activation='linear')) #Camada de saída para regressão linear

#compilando o modelo
model.compile(optimizer=Adam(learning_rate=0.1), loss='mean_squared_error')

#Treinando modelo
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)
# X_train: dados de entrada | y_train: respostas corretas | epochs=100: repete o treino 100 vezes | batch_size=32: treina em blocos de 32 exemplos | validation_split=0.2: usa 20% dos dados para validação | verbose=1: mostra o progresso na tela

#Fazendo previsões
y_pred = model.predict(X_test).flatten()

#Avaliação do modelo
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 =r2_score(y_test, y_pred)
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"R2: {r2}")

#Plotando os resultados
plt.scatter(y_test, y_pred, color ='LightSeaGreen')
plt.xlabel("Preço real")
plt.ylabel("Preço previsto")
plt.title("Preço real vs Preço previsto")
plt.show()

#plotando loss e val_loss
history = model.history.history
plt.plot(history['loss'], label='loss')
plt.plot(history['val_loss'], label='val_loss')
plt.xlabel("Épocas")
plt.ylabel("Loss")
plt.title("Loss e Val_loss durante o treinamento")
plt.legend()
plt.show()
```

```python
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

def build_model(hp):
    model = Sequential()

    # Camada de entrada
    model.add(Dense(
        units=hp.Int('units_input', min_value=32, max_value=512, step=32),
        activation='relu',
        input_shape=(X_train.shape[1],)
    ))

    # Camadas ocultas
    for i in range(hp.Int('num_layers', 1, 4)):
        model.add(Dense(
            units=hp.Int(f'units_{i}', min_value=16, max_value=256, step=16),
            activation='relu'
        ))
        if hp.Boolean(f'use_dropout_{i}'):
            model.add(Dropout(
                rate=hp.Float(f'dropout_rate_{i}', min_value=0.0, max_value=0.5, step=0.1)
            ))

    # Camada de saída
    model.add(Dense(1, activation='linear'))

    # Compilação
    model.compile(
        optimizer=Adam(
            learning_rate=hp.Float('learning_rate', min_value=1e-4, max_value=1e-1, sampling='LOG')
        ),
        loss='mean_squared_error'
    )

    return model
```

```python
# =====================================================
#   IMPORTS
# =====================================================
import os, sys, shutil, stat, random
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# =====================================================
#   REPRODUTIBILIDADE
# =====================================================
tf.keras.backend.clear_session()
os.environ["TF_DETERMINISTIC_OPS"] = "1"
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.keras.utils.set_random_seed(SEED)

# =====================================================
#   FUNÇÃO AUXILIAR: limpeza segura do diretório
#   (lida com permissão/encoding no Windows)
# =====================================================
def _handle_remove_readonly(func, path, exc_info):
    # Força permissão de escrita e tenta de novo
    os.chmod(path, stat.S_IWRITE)
    func(path)

def safe_clean_dir(path):
    """Remove arquivo/pasta no 'path' se existir. Ignora erros de encoding."""
    if os.path.isfile(path):
        try:
            os.remove(path)
        except Exception as e:
            print(f"[warn] não consegui remover arquivo {path}: {e}")
    elif os.path.isdir(path):
        try:
            shutil.rmtree(path, onerror=_handle_remove_readonly)
        except Exception as e:
            print(f"[warn] não consegui remover pasta {path}: {e}")

# =====================================================
#   ESCOLHA DE DIRETÓRIO 100% ASCII (evita UnicodeDecodeError)
#   -> evite acentos no caminho completo
# =====================================================
# Dica: se seu usuário do Windows tem acento, prefira uma raiz neutra:
if sys.platform.startswith("win"):
    base_dir = r"C:\kt_runs_v1"   # caminho simples, sem acentos
else:
    base_dir = os.path.abspath("kt_runs_v1")  # em UNIX, cwd costuma ser seguro

project_name = "house_price_v1"  # nome ASCII
safe_clean_dir(base_dir)         # limpa eventual resíduo anterior
os.makedirs(base_dir, exist_ok=True)

# =====================================================
#   DEFINIÇÃO DO MODELO P/ TUNER
# =====================================================
def build_model(hp):
    model = Sequential()

    # camada de entrada
    model.add(Dense(
        units=hp.Int('units_input', min_value=32, max_value=512, step=32),
        activation='relu',
        input_shape=(X_train.shape[1],)
    ))

    # camadas ocultas
    for i in range(hp.Int('num_layers', 1, 4)):
        model.add(Dense(
            units=hp.Int(f'units_{i}', min_value=16, max_value=256, step=16),
            activation='relu'
        ))
        if hp.Boolean(f'use_dropout_{i}'):
            model.add(Dropout(
                rate=hp.Float(f'dropout_rate_{i}', min_value=0.0, max_value=0.5, step=0.1)
            ))

    # saída
    model.add(Dense(1, activation='linear'))

    # compilação
    model.compile(
        optimizer=Adam(
            learning_rate=hp.Float('learning_rate', min_value=1e-4, max_value=1e-1, sampling='LOG')
        ),
        loss='mean_squared_error'
    )
    return model

# =====================================================
#   CONFIGURAÇÃO E BUSCA DO KERAS TUNER
# =====================================================
tuner = kt.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=10,
    executions_per_trial=2,
    directory=base_dir,          # ✅ diretório seguro (ASCII)
    project_name=project_name,   # ✅ projeto ASCII
    overwrite=True,              # ✅ sempre sobrescreve
    seed=42
)

tuner.search(X_train, y_train, epochs=100, validation_split=0.25)

# =====================================================
#   MELHOR MODELO + AVALIAÇÃO
# =====================================================
best_model = tuner.get_best_models(num_models=1)[0]
y_pred_best = best_model.predict(X_test).ravel()   # ✅ 1D

mae_best = mean_absolute_error(y_test, y_pred_best)
mse_best = mean_squared_error(y_test, y_pred_best)
r2_best  = r2_score(y_test, y_pred_best)

print(f"Melhor modelo - MAE: {mae_best:.4f}")
print(f"Melhor modelo - MSE: {mse_best:.4f}")
print(f"Melhor modelo - R2:  {r2_best:.4f}")

# =====================================================
#   PLOTS
# =====================================================
plt.scatter(y_test, y_pred_best, color='MediumPurple')
plt.xlabel("Preço real")
plt.ylabel("Preço previsto")
plt.title("Preço real vs Preço previsto - Melhor Modelo")
plt.show()

history = best_model.fit(X_train, y_train, epochs=100, validation_split=0.25, verbose=1).history
plt.plot(history['loss'], label='loss')
plt.plot(history['val_loss'], label='val_loss')
plt.xlabel("Épocas")
plt.ylabel("Loss")
plt.title("Loss e Val_loss durante o treinamento")
plt.legend()
plt.show()

# =====================================================
#   SALVAR E INSPECIONAR
# =====================================================
model_path = os.path.join(base_dir, "best_house_price_model_v1.h5")
best_model.save(model_path)
print(f"Modelo salvo em: {model_path}")

best_model.summary()
for layer in best_model.layers:
    if isinstance(layer, Dropout):
        print(f"Dropout rate: {layer.rate}")
```
